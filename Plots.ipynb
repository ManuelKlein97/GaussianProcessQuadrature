```python
W2 = np.zeros(np.shape(X2))
W3 = np.zeros(np.shape(X3))
W4 = np.zeros(np.shape(X4))
W5 = np.zeros(np.shape(X5))
W6 = np.zeros(np.shape(X6))
W7 = np.zeros(np.shape(X7))

for k in range(0, 99):
    kern = kernels.PolynomialBasis(input_dim=1, weight=b[k])
    W2[k] = functions.get_weights(kernel=kern, X=X2[k], A=[-1, 1])
    W3[k] = functions.get_weights(kernel=kern, X=X3[k], A=[-1, 1])
    W4[k] = functions.get_weights(kernel=kern, X=X4[k], A=[-1, 1])
    W5[k] = functions.get_weights(kernel=kern, X=X5[k], A=[-1, 1])
    W6[k] = functions.get_weights(kernel=kern, X=X6[k], A=[-1, 1])
    W7[k] = functions.get_weights(kernel=kern, X=X7[k], A=[-1, 1])
```
