```python
import numpy as np
import scipy as sp
import GPy
import functions
import kernels
from matplotlib import pyplot as plt
```

# Calculations

## Polynomial

```python
h = 0.01
NoP = int(round((1-h)/h, 0))
b = np.linspace(h, 1-h, NoP)
NoN = 7
NoT = 10
guesses = np.random.uniform(low=-0.75, high=0.75, size=(NoT, 2*NoN))
variance = np.ones((NoP, NoT))
X_optimal = np.ones((NoP, 2*NoN))
X_nodes = np.ones((NoP, NoT, 2*NoN))

for k in range(0, NoP):
    kern = kernels.PolynomialBasis(input_dim=1, weight=b[k])
    for i in range(0, NoT):
        X_nodes[k][i], variance[k][i] = functions.Optimizer_nodes_2D(kernel=kern, initialguess=guesses[i], return_var=True)
```

```python
for i in range(0, NoP):
    for k in range(0, NoT):
        X_optimal[i] = X_nodes[i][np.argmin(np.absolute(variance[i]))]
```

```python
X_optimal[98]
```

```python
ticks = np.array([-1, -0.5, 0, 0.5, 1])
index = 98
for i in range(0, 1):
    plt.figure()
    plt.grid()
    plt.plot(X_optimal[index][0:len(X_optimal):2], X_optimal[index][1:len(X_optimal):2], 'o')
    plt.ylim(-1, 1)
    plt.xlim(-1, 1)
    plt.xlabel('$x_1$', fontsize=15)
    plt.ylabel('$x_2$', fontsize=15)
    plt.gca().set_aspect('equal')
    plt.xticks(ticks, fontsize=14)
    plt.yticks(ticks,fontsize=14)
    plt.title('Number of nodes: ' + "{}".format(NoN), fontsize=15)
    plt.show()
```

```python
np.savetxt('./Nodes/' + 'OptimalNodes2DPBK_NoN_' + '{}'.format(NoN) + '.txt', X_optimal)
```

## Trigonometric

```python
R = np.zeros((6, 2*7))
```

```python
kern = kernels.Trigonometric(input_dim=1, weight=0.99)
for i in np.array([6, 7]):
    NoN = i
    guess = np.random.uniform(low=0.2, high=0.8, size=2*NoN - 2)
    temp = functions.Optimizer_nodes_2D(kernel=kern, initialguess=guess, return_var=False)
    for k in range(0, 2*NoN):
        R[i-2][k] = temp[k]
```
